```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys

support = 40
threshold = 5
```

```python
df = pd.read_csv('cleaned_data.csv', parse_dates=['BasketDate'])
```

```python
df = df[~df.BasketID.str.contains('C', na = False)]
```

```python
#df
```

```python
df.drop(columns=['BasketDate','Sale','Qta','ProdID','CustomerCountry'], inplace=True)
```

```python
#df
```

##Prefixspan

```python
series_of_all_carts = df.groupby(['CustomerID'])['BasketID'].unique().apply(list)
series_of_all_carts # get the series
```

```python
# get the list of carts for a customer having more than 10 carts 
list_of_all_carts = []
for list_of_carts_by_user in series_of_all_carts:
    if len(list_of_carts_by_user) > threshold:
        list_of_all_carts.append(list(list_of_carts_by_user))

len(list_of_all_carts)
```

```python
# list_of_all_carts
```

```python
import itertools
from prefixspan import PrefixSpan
```

```python
db = []

for list_of_customer_basket_id in list_of_all_carts:
    all_items_for_users = []
    for basketID in list_of_customer_basket_id:
        all_items_for_users.append(df[df['BasketID']== basketID]['ProdDescr'].values.tolist())
    all_items_for_users = list(itertools.chain(*all_items_for_users))
    db.append(all_items_for_users)
```

```python
ps = PrefixSpan(db)
```

```python
result = ps.frequent(support)
```

```python
result
```

```python
parameters = f'Support: {support}, Threshold: {threshold} \n'
print_result = f'{result} \n'
# Open a file with access mode 'a'
with open("results.txt", "a") as file_object:
    # Append 'hello' at the end of file
    file_object.write(parameters)
    file_object.write(print_result)
```
